1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
from scipy.stats import norm

# Parmetros da distribuio normal
media = 15.0
desvio_padrao = 0.1

# Peso limit
peso_limite = 15.2

# PROB
probabilidade = norm.sf(peso_limite, loc=media, scale=desvio_padrao)

print(f"A probabilidade de um pacote pesar mais que {peso_limite} kg  {probabilidade:.4f}")
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
from scipy.stats import norm

# Parmetros da distribuio normal
media = 15.0
desvio_padrao = 0.1

# P
percentil = 0.95

# Calcula 95%
peso_percentil_95 = norm.ppf(percentil, loc=media, scale=desvio_padrao)

print(f"O peso correspondente ao percentil de 95%  {peso_percentil_95:.4f} kg")
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
valores =np.linspace(-6, 6, 200)
import numpy as np
from scipy.stats import norm, t

# DNP
prob_normal = norm.cdf(valores, loc=0, scale=1)

# t-Student com 5
prob_t_student = t.cdf(valores, df=5)


for i, v in enumerate(valores):
    print(f"Valor: {v}")
    print(f"  P(N < {v}) = {prob_normal[i]:.4f}")
    print(f"  P(T < {v}) = {prob_t_student[i]:.4f}")
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python

import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(valores, prob_normal, label="Normal Padro (N(0,1))", color="blue", lw=2)
plt.plot(valores, prob_t_student, label="t-Student (df=5)", color="red", linestyle="--", lw=2)

# Config
plt.title("Comparao da CDF: Normal Padro vs. t-Student", fontsize=14)
plt.xlabel("Valor", fontsize=12)
plt.ylabel("Probabilidade Acumulada (CDF)", fontsize=12)
plt.axhline(0.5, color='gray', linestyle=':', lw=1)  # Linha para P=0.5
plt.axvline(0, color='gray', linestyle=':', lw=1)   # Linha para x=0
plt.legend(fontsize=12)
plt.grid(alpha=0.3)


plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
P =np.linspace(0, 1, 20)
# Seu cdigo comea aqui
import numpy as np
from scipy.stats import norm, t

# Percentis
percentis_normal = norm.ppf(P, loc=0, scale=1)

# Percentis t(5)
percentis_t_student = t.ppf(P, df=5)


for i, p in enumerate(P):
    print(f"Percentil: {p}")
    print(f"  Valor Normal Padro: {percentis_normal[i]:.4f}")
    print(f"  Valor t-Student (df=5): {percentis_t_student[i]:.4f}")
```

6. Faa um grfico com os dados do item anterior.

```python

# Create
plt.figure(figsize=(8, 6))
plt.plot(P, percentis_normal, marker='o', label="Normal Padro (N(0,1))", color="blue", lw=2)
plt.plot(P, percentis_t_student, marker='s', label="t-Student (df=5)", color="red", linestyle="--", lw=2)

# Config
plt.title("Comparao dos Percentis: Normal Padro vs. t-Student", fontsize=14)
plt.xlabel("Percentil (Probabilidade)", fontsize=12)
plt.ylabel("Valor Correspondente", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

plt.show()
```
